- (1 pts) Как связаны между собой задачи минимизации $f$ и $g$?
- (3 pts) Сравните градиентный метод и метод Ньютона для решения задач минимизации $f$ и $g$. Как связаны направления, получаемые каждым из методов? Как связаны методы в случае использования наискорейшего спуска для выбора шага?

### Решение

#### Пункт 1

Пусть $x^*$ --- точка минимума функции $f$. Тогда $\forall x\in \mathbb{R}\ \ f(x) \geq f(x^*)$. В силу того что $\varphi$ возрастает, то $\varphi(f(x)) \geq \varphi(f(x^*))$. Следовательно, точки минимума функций $f$ и $g$ совпадают.

#### Пункт 2

$$g'(x) = (\varphi(f(x)))' = \varphi'(f(x)) f'(x)$$

В силу того, что $\varphi$ возрастает, имеем для любого $x$ выполнено $\varphi'(f(x)) > 0$. Значит, в любой точке градиенты функций $f$ и $g$ направлены в одну сторону и отличаются лишь длиной: $\frac{\|g'(x)\|}{\|f'(x)\|} = \varphi'(f(x))$ .

## Задача 3 (4 pts)
В этой задаче надо вывести формулу Шермана-Моррисона-Вудберри

- (2 pts) Дано $A \in \mathbb{R}^{m \times n}$, $B \in \mathbb{R}^{n \times m}$ и $I + AB$ обратима. Покажите, что и $I + BA$ обратима, а также, что $(I + BA)^{-1}B = B(I + AB)^{-1}$ 
- (0.5 pts) Выведите, что $(I+P)^{-1} = I - (I+P)^{-1}P = I - P(I+P)^{-1}$
- (0.5 pts) Скомбинируйте два предыдущих факта и покажите, что $(I + UV)^{-1} = I - UV(I + UV)^{-1} = I - U(I + VU)^{-1}V$
- (1 pts) Обобщите предудыщий факт для матрицы $A + UV$ и получите формулу Шермана-Моррисона-Вудбери

$$
(A + UV)^{-1} = A^{-1} - A^{-1} U \left( I_k + V A^{-1} U \right)^{-1} V A^{-1},
$$

где $A \in \mathbb{R}^{n \times n}$, $U \in \mathbb{R}^{n \times k}$ и $V \in \mathbb{R}^{k \times n}$.


### Решение

#### Пункт 1

Покажем, что матрица $I - B(I+AB)^{-1}A$ и есть обратная к $I + BA$. 

$$(I + BA)(I - B(I+AB)^{-1}A) = I + BA - B(I+AB)^{-1}A - BAB(I+AB)^{-1}A = I + BA - B((I+AB)^{-1} + AB(I+AB)^{-1})A = I + BA - B((I+AB)^{-1})(I+AB))A =  I + BA - B I A = I$$

Отсюда и следует, что $(I + BA)^{-1} = I - B(I+AB)^{-1}A$.





Пусть $(I+BA)$ вырождена. Тогда существует такой ненулевой вектор $x$, что $(I+BA)x = 0$, $BAx = -x$. Домножим обе части этого равенства на $A$ и обо значим $y = Ax$.  

$$ABAx = -Ax$$

$$ABy = -y$$

$$(I+AB)y=0$$

Заметим, что если $y = Ax = 0$, то $(I+BA)x = x + BAx = x + B \times 0 = x \neq 0$, поэтому $y$ ненулевой, и  у матрицы $(I+AB)y=0$ есть собственное значение $0$, а следовательно она вырождена. Но по условию $(I+AB)$ имеет обратную. Значит, наше изначальное предположение неверно и тоже $(I+BA)$ невырождена.

Осталось разобраться с тождеством $(I + BA)^{-1}B = B(I + AB)^{-1}$. Домножим его слева на $(I+BA)$, а справа на $(I+AB)$. Получим

$$B(I+AB)=(I+BA)B$$

Раскрыв скобки увидим, что с обеих сторон равенства стоит $B+BAB$. То есть, это действительно тождество.

#### Пункт 2

$$(I+P)(I - (I+P)^{-1}P ) = I+P - (I+P)(I+P)^{-1}P = I+P-P = I \Rightarrow (I+P)^{-1} = I - (I+P)^{-1}P $$

$$(I - P(I+P)^{-1}) (I+P) = I+P - P(I+P)^{-1}(I+P) = I+P-P = I \Rightarrow (I+P)^{-1} = I - P(I+P)^{-1}$$

#### Пункт 3

Подставив $P+UV$ в формулу из пункта 2, получим $(I + UV)^{-1} = I - UV(I + UV)^{-1}$. А затем применим тождество $B(I+AB)=(I+BA)B$ из пункта 1 и получим $I - UV(I + UV)^{-1} = I - U(I + VU)^{-1}V$.
