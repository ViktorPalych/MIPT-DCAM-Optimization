### Пункт 1

$$f^*(y) = \sup_x (yx - |x|)$$

1) $x \geq 0$. Тогда $yx - |x| = (y-1)x$. 

В этом случае если $y > 1$, устремляя $x \rightarrow +\infty$, Получаем $ (y-1)x\rightarrow +\infty$ и $f^*(y) = +\infty$. 

А если же $y \leq 1$, то $(y-1)x \leq 0$.

2) $x < 0$. Тогда $yx - |x| = (y+1)x$. 

Здесь уже если $y < -1$, то при $x \rightarrow -\infty$, Получаем $ (y+1)x\rightarrow +\infty$ и $f^*(y) = +\infty$. 

А в случае когда $y \geq -1$ имеем $(y+1)x \leq 0$.

Собрав все эти факты, запишем выражение 

\begin{align}

\begin{cases}

\end{align}

В матрице $A$ строк больше чем столбцов. Рассмотрите две различных функции, сглаживающие $\|x\|_1$. Покажите, как для каждой из них сходимость зависит от параметра сглаживания.

````python
import numpy as np

def prox_alg(f, f_grad, g_prox, x0, num_iter, beta=0.5, fix_lam=None, accel=False):
    conv = [x0]
    x = x0.copy()
    if accel:
        t_prev = 1
        t_next = (1 + np.sqrt(1 + 4 * t_prev**2)) / 2.
    if fix_lam is None:
        lam = 1.
    for i in range(num_iter):
        if accel and i > 0:
            x = x + (t_prev - 1) / t_next * (x - conv[-2])
            t_prev = t_next
            t_next = (1 + np.sqrt(1 + 4 * t_prev**2)) / 2.
        if fix_lam is None:
            while True:
                z = g_prox(x - lam * f_grad(x), lam)
                if f(z) <= f(x) + f_grad(x).dot(z - x) + np.linalg.norm(z - x)**2 / (2 * lam):
                    break
                else:
                    lam = lam * beta
        else:
            z = g_prox(x - fix_lam * f_grad(x), fix_lam)
        x = z.copy()
        conv.append(x)
    return x, conv
````

````python
import liboptpy.unconstr_solvers as methods
import liboptpy.step_size as ss

def huber(x):
    res = np.zeros_like(x)
    res[np.abs(x) <= mu] = x[np.abs(x) <= mu]**2 / (2 * mu)
    res[np.abs(x) >= mu] = np.abs(x[np.abs(x) >= mu]) - mu / 2
    return np.linalg.norm(A.dot(x) - b)**2 / 2 + gamma * np.sum(res)

def grad_huber(x):
    res = np.zeros_like(x)
    res[np.abs(x) <= mu] = x[np.abs(x) <= mu] / mu
    res[np.abs(x) >= mu] = np.sign(x[np.abs(x) >= mu])
    return A.T.dot(A.dot(x) - b) + gamma * res

def sqrt_approx(x):
    return np.linalg.norm(A.dot(x) - b)**2 / 2 + gamma * np.sum([(np.sqrt(xi**2+mu**2) - mu) for xi in x])

def grad_sqrt(x):
    return A.T.dot(A.dot(x) - b) + gamma * np.array([xi / np.sqrt(xi**2+mu**2) for xi in x])
````
